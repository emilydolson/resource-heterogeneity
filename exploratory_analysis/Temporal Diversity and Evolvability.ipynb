```python
import os
import numpy, scipy, matplotlib
import matplotlib.pyplot as plt
```

```python
def getDataEquTime(condition, taskName='equ'):
    diversities = []
    task = []
    try:
        dirs = os.listdir(condition)
    except:
        print condition, "not found"
        return [], []
    for d in dirs:
        taskfile = open(condition + "/" + d + "/tasks.dat")
        taskConts = taskfile.readlines()
        taskInd = {}
        indices = [line.strip("# ").split(":") for line in taskConts[5:14]]
        for line in indices:
            name = line[1].strip().lower()
            if name != "nand" and name != "andnot":
                name = name[:3]
            elif name == "andnot":
                name = "andn"
            taskInd[name] = int(line[0].strip()) - 1

        taskConts = taskConts[15:]
        diversityfile = open(condition + "/" + d +"/phenotype_count.dat")
        divData = diversityfile.readlines()[8:]
        for  i in range(len(taskConts)):
            if int(taskConts[i].split()[taskInd[taskName]]) > 0:
                task.append(float(taskConts[i].split()[0]))
                diversities.append(float(divData[i].split()[2]))
                break
        taskfile.close()
        diversityfile.close()
        
    return diversities, task

def doStatsEquTime(conditions, task="equ"):
    divs = []
    equs = []
    for c in conditions:
        d, e = getDataEquTime(c, task)
        if d == []:
            print "skipping", c
            continue
        divs.append(d)
        equs.append(e)
    
    plt.figure(1)
    #plt.axes().set_xticks([i+.5 for i in range(len(equHeights))])
    inflow = True
    for i in range(len(conditions)):
        if conditions[i][:3] == "rad":
            inflow = False
            break
        if conditions[i][:14] == "max_background":
            conditions[i] = "6"
    plt.axes().set_xticklabels([c.strip("inflowradius_limEQqU") for c in conditions])
    if inflow:
        plt.xlabel("Rinflow")
        plt.title("Diversity Distributions over Resource Inflow Rates")
    else:
        plt.xlabel("Patch radius")
        plt.title("Diversity Distributions over Patch Sizes")
    plt.ylabel("Phenotypic Shannon entropy of viable organisms")

    plt.boxplot(divs, notch=True)
    plt.ylim(0, 6)
    
    stdDev = [sqrt(float(sum(e))/float(len(e))) for e in equs]
    avgs = [float(sum(e))/float(len(e)) for e in equs]    

    plt.figure(2)
    plt.xlim(0,len(avgs))
    if inflow:
        plt.title("Evolution of " + task + " over Resource Inflow Rates")
        plt.xlabel("Rinflow")
    else:
        plt.title("Evolution of " + task + " over Resource Patch Sizes")
        plt.xlabel("Patch radius")
    plt.ylabel("# of populations that evolved " + task + " by 100,000 updates")
    

    print [len(e) for e in equs]
    plt.axes().set_xticks([i+.5 for i in range(len(avgs))])
    for i in range(len(conditions)):
        if conditions[i][:14] == "max_background":
            conditions[i] = "6"
    plt.axes().set_xticklabels([c.strip("inflowradius_limEQqU") for c in conditions])
    plt.bar(range(len(avgs)), avgs, yerr=stdDev, ecolor="black")
```
