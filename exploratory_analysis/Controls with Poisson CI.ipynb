```python
import os
import numpy, scipy, matplotlib
import matplotlib.pyplot as plt
import csv
import sys

path = "/user_home/w_BEACONGroupE/resource-heterogeneity/control/"
os.chdir(path)
conditions = ["inflow1redo", "inflow3redo", "inflow10redo", "inflow30", "inflow100", "inflow300", "inflow1000", "inflow3000", "inflowInfinite"]

def getData(condition):
    diversities = []
    equs = []
    try:
        dirs = os.listdir(condition)
    except:
        return [], []
    for d in dirs:
        diversityfile = open(condition + "/" + d +"/phenotype_count.dat")
        diversities.append(float(diversityfile.readlines()[-1].split()[2]))
        diversityfile.close()
        taskfile = open(condition + "/" + d + "/tasks.dat")
        equs.append(int(taskfile.readlines()[-1].split()[9]))
        taskfile.close()
   
    #print equs
    csv_out = open('control_equs.csv', 'a')
    mywriter = csv.writer(csv_out)
    
    for j in range(len(equs)):
        mywriter.writerow([condition,equs[j]])
    csv_out.close()
    
    #print diversities
    csv_out2 = open('control_divs.csv', 'a')
    mywriter2 = csv.writer(csv_out2)
    
    for j in range(len(diversities)):
        mywriter2.writerow([condition,diversities[j]])
    csv_out2.close()
    
    
    return diversities, equs


def doStats(conditions):
    divs = []
    equs = []
    for c in conditions:
        d, e = getData(c)
        if d == []:
            continue
        divs.append(d)
        e = [int(i>0) for i in e]
        equs.append(e)
    
    plt.figure(1)
    plt.xlabel("Rinflow")
    plt.ylabel("Phenotypic Shannon entropy of viable organisms")
    plt.title("Diversity Distributions over Resource Inflow Rates")
    plt.boxplot(divs,notch=True)
    equHeights = [sum(i) for i in equs]
    plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], ['1', '3', '10', '30', '100', '300', '1000', '3000', 'Infinite'], rotation='horizontal')
    
    
    plt.figure(2)
    plt.title("Evolution of EQU over Resource Inflow Rates")
    plt.xlabel("Rinflow")
    plt.ylabel("# of populations that evolved EQU by 100,000 updates")
    log10stderror=[84674.823,84674.823,0.3015113,0.2041241,0.2182179,0.1961161,0.2886751,0.4472136,0.4082483]
    stderror=[84674.823,84674.823,2.002217717,1.600015169,1.65279085,1.570782666,1.943905283,2.800358287,2.56004913]
    plt.bar(range(len(equHeights)), equHeights, yerr=stderror)
    plt.xticks([0.4, 1.4, 2.4, 3.4, 4.4, 5.4, 6.4, 7.4, 8.4], ['1', '3', '10', '30', '100', '300', '1000', '3000', 'Infinite'], rotation='horizontal')
    plt.ylim(0,30)
    
        
doStats(conditions)
getData(conditions)

```
