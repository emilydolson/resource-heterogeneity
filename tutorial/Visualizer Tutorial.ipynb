To visualize spatial phenotype placement, use the makeMostCommonGrid() function in genHeterogenousEnvironment2. As arguments, it takes a list of filenames (as strings), and a string indicating the mode. This string can either be "species" or "ntasks". "species" will color each phenotype differently, while "ntasks" will just only color phenotypes that can perform different numbers of tasks differently (making it useful for aggregating across treatments in which the resources at a set of fixed anchor poitns were randomized). In either case, colors function as a heat map, with cooler colors indicating more complex phenotypes. "-intensify" can be appended to either mode to multiply the color values by a factor to bring out the differentiation between them. Aggregation is done by taking the most common phenotype or number of tasks in a given cell across all files in the input list.

As a backwards compatability features, makeMostCommonGrid() returns an ipython blocks grid and an array of data. The ipython blocks grid can be visualized by executing the name of variable it is stored in in an ipython notebook. The data array can be used with the more versatile imshow function (from matplotlib) to generate the heat map as a standard image.

```python
files = glob.glob("experiment/2circles_notXorn21/*/grid_task.100000.dat")
grid, data = makeMostCommonGrid(files, "species-intensify")
figure(figsize=(8,8))
imshow(data, interpolation="none")
```

If blobs of similar phenotypes occupying a certain niche are less important than the average number of tasks being done in various locations, you can use the makeAggregateNTasksGrid() function. It works much the same as makeMostCommonGrid(), but instead plots the mean appropriate color value rather than the mode. Also, it can only plot by number of tasks (since averaging phenotypes wouldn't entirely make sense). Setting the second argument to True intensifies colors in the same way that the "-intensify" flag does in makeAggregateNTasksGrid().

```python
files = glob.glob("experiment/2circles_notXorn14/*/grid_task.100000.dat")
grid, data = makeAggregateNTasksGrid(files[:], True)
figure(figsize=(8,8))
imshow(data, interpolation="none")
```

To visualize the placement of resources in an environment, use the calcGrid() function. calcGrid takes a list of resource names, a list of anchor points (tuples representing x,y coordinates), and a radius and displays the placement of resources in the world using ipython blocks. The list of resources can be generated at random using genRandResources(), which uses args.patchesPerSide to calculate the number of resources to place and the global list of all possible resource types to select a roughly even number of each resource. Anchor points can be automatically generated using any of the anchor generation functions: calcEvenAnchors() (calculates evenly placed anchors), calcTightAnchors()(takes distance as an argument and returns anchor points with that number of cells in between neighboring circles), or calcRandAnchors() (generates random anchors that are far enough from the edges to keep circles from being cut off). All anchor generation functions automatically generate the correct number of anchor points.

```python
randResources = genRandResources()
anchors = calcEvenAnchors()
grid = calcGrid(randResources, anchors, 24)
grid
```

Phenotypes will never, in practice, be constrained to the tight boundaries of spatial niches. However, it is still sometimes useful (and pretty!) to visualize the theoretical shapes of the niches. The most effective way to do this (because it generates across randomized resource types) is by generating a heat map of the number of tasks rewarded in each location. This can be accomplished with calcNResGrid, which works much the same way as calcGrid(), aside from the aforementioned difference in the quantity being visualized.

```python
randResources = genRandResources()
anchors = calcEvenAnchors()
grid = calcNResGrid(randResources, anchors, 24)
grid
```
